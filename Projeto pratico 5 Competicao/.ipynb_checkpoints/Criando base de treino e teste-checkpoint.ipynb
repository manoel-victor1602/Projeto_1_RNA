```python
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_csv('dataR2.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_data.head()
```

```python
df_data.BMI = df_data.BMI.apply(round, args=[4])
df_data.HOMA = df_data.HOMA.apply(round, args=[4])
df_data.Adiponectin = df_data.Adiponectin.apply(round, args=[4])
df_data.Resistin = df_data.Resistin.apply(round, args=[4])
```

```python
df_data['Soma'] = [np.sum(d) for d in df_data.drop('Classification', axis=1).values]
```

```python
df_test['Soma'] = [np.sum(t) for t in df_test.drop('id', axis=1).values]
```

```python
df_data[df_data.Age == 29]
```

```python
df_test
```

```python
test_ids = []
for t in df_test.Soma:
    cont = 0
    for d in df_data.Soma:
        if(t == d):
            test_ids.append(cont)
        cont += 1
test_ids
```
